```python
tips = sns.load_dataset('tips')
```

```python
tips.head(2)
```

```python
tips.info()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.hist(tips['total_bill'], bins=10) # bins => 막대바개수 기본값은 10
ax1.set_title('Histogram of Total Bill')
ax1.set_xlabel('Frequeny')
ax1.set_ylabel('total bill')
plt.show()
```

```python

scatter_plot = plt.figure()
ax1 = scatter_plot.add_subplot(111) # ,(콤마) 없이 써도 ㄱㅊ
ax1.scatter(tips['total_bill'], tips['tip'])
plt.show()
```

```python
# boxplot
boxplot = plt.figure()
ax1 = boxplot.add_subplot(111)
data = [tips[tips['sex'] == 'Female']['tip'], tips[tips['sex'] == 'Male']['tip']]
ax1.boxplot(data, labels=['Famale','Male'])
plt.show()
```

```python
[tips[tips['sex'] == 'Female']['tip'], tips[tips['sex'] == 'Male']['tip']]
```

```python
def recode_sex(sex):
    if sex == 'Female':
        return 0
    else:
        return 1
```

```python
tips['color'] = tips['sex'].apply(recode_sex)
```

```python
tips
```

```python
scatter_plot = plt.figure()
ax1 = scatter_plot.add_subplot(111)
ax1.scatter(tips['total_bill'],
             tips['tip'],
            s = tips['size']*10,
            c = tips['color'],
            alpha = 0.5 )
ax1.set_title('산점도') 
```

```python
import seaborn as sns
tips = sns.load_dataset('tips')
```

```python
# sns.distplot(tips['total_bill'])
ax = sns.histplot(data = tips, x='total_bill', bins=8, kde=True)
ax.set_title('히스토그램')
```

```python
sns.__version__
```

```python
ax = sns.countplot(data=tips, x = 'day')
ax.set_title('count of days')
ax.set_xlabel('day')
ax.set_ylabel('빈도')
```

```python
sns.regplot(data = tips, x = 'total_bill', y = 'tip', fit_reg = False)
```
